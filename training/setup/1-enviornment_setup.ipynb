## MLflow Tracking: Three Deployment Patterns

Before we set up the local environment, it's worth understanding the three ways MLflow can be deployed — so you know where this workshop fits in the bigger picture.

![MLflow server deployment patterns](../assets/tracking-setup-overview.png)

| Pattern | How it works | When to use it |
|---------|-------------|----------------|
| **Local (filesystem)** | MLflow writes directly to `./mlruns` on disk | Quick prototyping; being deprecated in MLflow 3 (Deprecated) |
| **Local server + SQLite** | `mlflow server` runs locally; data stored in a `.db` file | ✅ **This workshop** — full features, no cloud needed |
| **Remote server** | Shared MLflow instance (Databricks, self-hosted, etc.) | Teams & production workloads |

For this workshop we use **pattern 2**: a local MLflow server backed by SQLite. This gives you the full MLflow feature set (tracing, evaluation, prompt registry) without needing any cloud credentials.

### Step 2 — Install pyenv (optional but recommended)

**pyenv** lets you install and switch between multiple Python versions without touching the system Python. This is especially useful when different projects require different Python versions.

Without pyenv, you can still follow along as long as you have Python 3.9+ installed (required by mlflow) — skip to Step 4.

Install pyenv via Homebrew (macOS):

> **Windows users:** use [pyenv-win](https://github.com/pyenv-win/pyenv-win) instead.

````python
brew update
brew install pyenv

# Add pyenv to your shell — paste this into ~/.zshrc or ~/.bashrc, then restart your terminal:
# export PYENV_ROOT="$HOME/.pyenv"
# export PATH="$PYENV_ROOT/bin:$PATH"
# eval "$(pyenv init -)"
````

### Step 3 — Install and pin a Python version

`pyenv install` downloads the interpreter; `pyenv local` writes a `.python-version` file into your project directory so every tool (including `uv`) automatically picks up the right version.

````python
# Install a specific Python version to pyenv (makes it available for local install)
pyenv install 3.14

# Pin it to this project directory (creates a .python-version file)
pyenv local 3.14

# Set global python version (different from local)
pyenv global 3.13

# Verify
python --version
````

### Step 4 — Install uv

**uv** is a next-generation Python package manager written in Rust. It replaces `pip`, `venv`, and `pip-tools` in a single tool and is significantly faster — cold installs that take 30 seconds with pip typically take under 2 seconds with uv.

Key things uv manages for you:
- Creates and manages `.venv` (the virtual environment)
- Tracks exact dependency versions in `uv.lock` (like `package-lock.json`)
- Reads project metadata from `pyproject.toml`

````python
# Install uv globally (only needs to be done once per machine)
pip install uv

# Or on macOS with Homebrew:
# brew install uv

# Verify
uv --version
````

### Step 6 — Initialize the uv project

`uv init` creates the core project scaffolding:
- `pyproject.toml` — project metadata and dependency list
- `.venv/` — the isolated virtual environment
- `uv.lock` — pinned dependency tree (commit this to git)

If you cloned the repo, `uv init` has already been run — just run `uv sync` to install the existing dependencies instead.

````python
# If you're starting from scratch inside an existing empty folder
uv init

# If you have a .toml file
uv sync

# If you want uv to create a new subfolder for you
# uv init my-project-name
````

````python
# If something goes wrong (corrupt .venv, version conflicts, etc.)
# uv sync --reinstall rebuilds the environment from scratch using uv.lock
uv sync --reinstall
````

### Step 7 — Activate the virtual environment

Activating the venv tells your terminal to use the Python and packages inside `.venv/` instead of your global Python. You need to do this **in every new terminal session**.

> **VS Code tip:** If you open VS Code from within your project folder, it will usually detect and activate the venv automatically when you select the right Python interpreter (⌘⇧P → "Python: Select Interpreter" → choose `.venv`).

````python
# macOS / Linux
source .venv/bin/activate

# Windows (PowerShell)
# .venv\Scripts\Activate.ps1

# Confirm the right Python is active — path should include .venv
which python
````

### Step 8 — Add ipykernel

`ipykernel` is the bridge between your virtual environment and Jupyter. Without it, VS Code's notebook runner can't see the packages installed in your `.venv` — it would fall back to the global Python.

After running this, VS Code will show your project's venv as a selectable kernel in the notebook kernel picker (top-right corner of any `.ipynb` file).

````python
uv add ipykernel
````

### Step 9 — Verify everything is working

Run this cell inside the notebook (not in the terminal) using your project's kernel. If the kernel picker (top-right of the notebook) shows your `.venv`, you're good. If it shows a system Python, switch it first.

````python
import sys

print(f"Python version : {sys.version}")
print(f"Python location: {sys.executable}")

# The path should contain '.venv' — if it points to a system Python, switch your kernel
if ".venv" in sys.executable:
    print("\n✅ Running inside the project virtual environment")
else:
    print("\n⚠️  Not using the project venv — select the right kernel (top-right of the notebook)")
````
